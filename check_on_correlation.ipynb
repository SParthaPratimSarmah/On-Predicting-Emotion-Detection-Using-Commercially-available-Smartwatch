```python
emotion_columns = ['ALERT', 'ANGER', 'FEAR', 'HAPPY', 'NEUTRAL', 'RESTED/RELAXED',
                   'SAD', 'SURPRISE', 'TENSE/ANXIOUS', 'TIRED']

# Create a new column 'Emotions'
df['Emotions'] = df[emotion_columns].idxmax(axis=1)

# If you want to replace NaN values with a default value, you can use:
df['Emotions'].fillna('No Emotion', inplace=True)
df
```
