```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('ML_algorithm/data/encoded_two_users.csv')

# Combine emotion columns into a single "Emotions" column
emotion_columns = ['ALERT_True', 'ANGER_True', 'FEAR_True', 'HAPPY_True', 'NEUTRAL_True',
                   'RESTED/RELAXED_True', 'SAD_True', 'SURPRISE_True', 'TENSE/ANXIOUS_True', 'TIRED_True']
df['Emotions'] = df[emotion_columns].idxmax(
    axis=1).apply(lambda x: x.replace('_True', ''))

# Drop the original emotion columns
df.drop(emotion_columns, axis=1, inplace=True)

# Optional: Drop the 'False' columns as well if they are no longer needed
false_columns = [col for col in df.columns if col.endswith('_False')]
df.drop(false_columns, axis=1, inplace=True)

# Save the DataFrame with the new "Emotions" column
df.to_csv(
    'ML_algorithm/data/path_to_save_updated_csv_file_two_users.csv', index=False)

# Display the first few rows to verify the changes
df.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Load the CSV file into a DataFrame
df = pd.read_csv('ML_algorithm/data/encoded_two_users.csv')

# Combine emotion columns into a single "Emotions" column
emotion_columns = ['ALERT_True', 'ANGER_True', 'FEAR_True', 'HAPPY_True', 'NEUTRAL_True',
                   'RESTED/RELAXED_True', 'SAD_True', 'SURPRISE_True', 'TENSE/ANXIOUS_True', 'TIRED_True']
df['Emotions'] = df[emotion_columns].idxmax(
    axis=1).apply(lambda x: x.replace('_True', ''))

# Drop the original emotion columns
df.drop(emotion_columns, axis=1, inplace=True)

# Optional: Drop the 'False' columns as well if they are no longer needed
false_columns = [col for col in df.columns if col.endswith('_False')]
df.drop(false_columns, axis=1, inplace=True)

# Select features and target
features = df.drop(['id', 'date', 'Emotions'], axis=1)
target = df['Emotions']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.2, random_state=42)

# Train a model (RandomForestClassifier used as an example)
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')

# Add predictions to the DataFrame
df['Predicted_Emotions'] = model.predict(features)

# Save the DataFrame with the new "Emotions" column
df.to_csv('path_to_save_updated_csv_file.csv', index=False)
```

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('ML_algorithm/data/fifiteen_users_data.csv')

# Combine emotion columns into a single "Emotions" column
emotion_columns = ['ALERT_True', 'HAPPY_True', 'NEUTRAL_True',
                   'RESTED/RELAXED_True', 'SAD_True', 'TENSE/ANXIOUS_True', 'TIRED_True']
df['Emotions'] = df[emotion_columns].idxmax(
    axis=1).apply(lambda x: x.replace('_True', ''))

# Drop the original emotion columns
df.drop(emotion_columns, axis=1, inplace=True)

# Optional: Drop the 'False' columns as well if they are no longer needed
false_columns = [col for col in df.columns if col.endswith('_False')]
df.drop(false_columns, axis=1, inplace=True)

# Save the DataFrame with the new "Emotions" column
df.to_csv('ML_algorithm/data/fifiteen_users_data_updated_csv_file.csv', index=False)

# Display the first few rows to verify the changes
df.head()
```
