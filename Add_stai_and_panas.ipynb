# Import stai


```python
stai = pd.read_csv("STAI.csv")
stai
```

```python
stai['Submit_Date'] = pd.to_datetime(stai['Submit_Date']).dt.date
stai
```

```python
stai = stai.drop(columns=['Start_Date', 'Date_Timestamp'])
stai
```

```python
# convert 5-likert to 4-likert scale (mistakenly the STAI scale was distributed with a 5-likert, but it's originally 4-likert)
def convert_5_to_4_likert(x):
    return (4 - 1) * (x - 1) / (5 - 1) + 1


stai.iloc[:, 3:] = stai.iloc[:, 3:].apply(
    lambda x: convert_5_to_4_likert(x))
stai
```

```python
def proper_round(num, dec=0):
    num = str(num)[:str(num).index('.')+dec+2]
    if num[-1] >= '5':
        a = num[:-2-(not dec)]       # integer part
        b = int(num[-2-(not dec)])+1  # decimal part
        return float(a)+b**(-dec+1) if a and b == 10 else float(a+str(b))
    return float(num[:-1])


for col in stai.iloc[:, 3:].columns:
    stai[col] = stai[col].apply(lambda x: proper_round(x))
stai
```

```python
def inverse_score(score, min, max):
    return max - score + min
```

```python
users_personality = pd.read_csv("../../../surveys/personality.csv")
```

```python
# Based on the scoring document: https://oml.eular.org/sysModules/obxOML/docs/id_150/State-Trait-Anxiety-Inventory.pdf, some questions are reversed in STAI
stai_reversed = ["Question_01", "Question_02]", "Question_05]", "Question_08]",
                 "Question_10]", "Question_11]", "Question_15]", "Question_16]", "Question_19]", "Question_20]"]
for col in users_personality.columns:
    # inversing scores for the minus keyed items; the plus keyed items stay as is
    if col in stai_reversed:
        stai[col] = stai[col].apply(
            lambda score: inverse_score(score, 1, 4))
stai
```

```python
stai['stai_stress'] = stai.iloc[:, 3:].sum(axis=1)
stai
```

```python
mean_stai = stai['stai_stress'].mean()
std_stai = stai['stai_stress'].std()


def get_stai_category(score):
    if score < mean_stai-0.5*std_stai:
        return "Below average"
    if score > mean_stai+0.5*std_stai:
        return "Above average"
    return "Average"


stai['stai_stress_category'] = stai['stai_stress'].apply(
    lambda score: get_stai_category(score))
stai
```

```python
# Ensure there are no extra spaces or tab characters in column names
stai.columns = stai.columns.str.strip()

# Drop duplicate rows based on the combination of 'user_id' and 'Submit_Date'
b = stai.drop_duplicates(
    subset=["user_id", "Submit_Date"], inplace=True, keep="last")

print(b)
# Print the number of survey responses after duplicates elimination
print(
    "We have {} survey responses for the STAI (Y1-form) scale after duplicates elimination.".format(stai.shape[0]))
```

```python
print("Average value on the S-STAI scale: {} - Standard deviation on the S-STAI scale: {}".format(mean_stai, std_stai))
```

## Filling all the missing data


```python
df = pd.read_csv("Updated_Fitbit_sema_survey_panas_stai.csv")
df
```

```python
df.info()
```

# Fillup categorical values with mood


```python
categorical_cols = ['badgeType', 'activityType', 'mindfulness_session', 'stai_stress_category',
                    'ALERT', 'HAPPY', 'NEUTRAL', 'RESTED/RELAXED', 'SAD', 'TENSE/ANXIOUS', 'TIRED']
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.info()
```

# Median imputation for numerical value


```python
numerical_cols = ['calories', 'steps', 'bpm', 'sleep_duration', 'minutesToFallAsleep', 'resting_hr',
                  'positive_affect_score', 'negative_affect_score', 'stai_stress']
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())
```

# Time-series columns with forward/backward fill


```python
time_series_cols = ['minutesToFallAsleep',
                    'minutesAsleep', 'minutesAwake', 'steps']
for col in time_series_cols:
    df[col] = df[col].fillna(
        method='ffill').fillna(method='bfill')
```

# survey data using knn


```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5)
affect_columns = ['positive_affect_score',
                  'negative_affect_score', 'stai_stress']
df[affect_columns] = imputer.fit_transform(df[affect_columns])
```

### demographic data with group based imputation


```python
df['age'] = df['age'].replace(
    {'<30': 20, '>=30': 30}).infer_objects(copy=False)
df.age
```

```python
df['bmi'] = df['bmi'].replace(
    {'<19': 18, '>=25': 25, '>=30': 30})
df['bmi'] = df['bmi'].infer_objects(copy=False)
```

```python
df['age'] = df.groupby(
    'gender')['age'].transform(lambda x: x.fillna(x.median()))
df['bmi'] = df.groupby(
    'gender')['bmi'].transform(lambda x: x.fillna(x.median()))
```

```python
df.info()
```

```python
# Impute missing numerical values with the median
numerical_cols = ['nightly_temperature', 'nremhr', 'rmssd', 'entropy', 'spo2',
                  'full_sleep_breathing_rate', 'full_sleep_standard_deviation', 'full_sleep_signal_to_noise',
                  'deep_sleep_breathing_rate', 'deep_sleep_standard_deviation', 'deep_sleep_signal_to_noise',
                  'light_sleep_breathing_rate', 'light_sleep_standard_deviation', 'light_sleep_signal_to_noise',
                  'rem_sleep_breathing_rate', 'rem_sleep_standard_deviation', 'rem_sleep_signal_to_noise',
                  'stress_score', 'sleep_points_percentage', 'exertion_points_percentage', 'responsiveness_points_percentage',
                  'daily_temperature_variation', 'filteredDemographicVO2Max', 'distance', 'lightly_active_minutes',
                  'moderately_active_minutes', 'very_active_minutes', 'sedentary_minutes', 'scl_avg', 'minutesAfterWakeup',
                  'sleep_efficiency', 'sleep_deep_ratio', 'sleep_wake_ratio', 'sleep_light_ratio', 'sleep_rem_ratio',
                  'minutes_in_default_zone_1', 'minutes_below_default_zone_1', 'minutes_in_default_zone_2', 'minutes_in_default_zone_3',
                  'min_goal', 'max_goal']

for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Fill missing boolean values with False
boolean_cols = ['HAPPY', 'NEUTRAL', 'RESTED/RELAXED',
                'SAD', 'TENSE/ANXIOUS', 'TIRED']

for col in boolean_cols:
    df[col] = df[col].fillna(False)
```

```python
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
df.info()
```

```python
# Impute age, gender, bmi with mode or median
df['age'] = df['age'].fillna(df['age'].mode()[0])  # Mode for categorical
df['gender'] = df['gender'].fillna(
    df['gender'].mode()[0])  # Mode for categorical
df['bmi'] = df['bmi'].fillna(df['bmi'].median())  # Median for numerical

# Impute step_goal, step_goal_label, <no-response>, ANGER, FEAR, etc. with mode
categorical_cols_with_high_na = ['step_goal', 'step_goal_label', '<no-response>',
                                 'ANGER', 'FEAR', 'SURPRISE', 'ENTERTAINMENT',
                                 'GYM', 'HOME', 'HOME_OFFICE', 'OTHER',
                                 'OUTDOORS', 'TRANSIT', 'WORK/SCHOOL']

for col in categorical_cols_with_high_na:
    if df[col].dtype == 'object':  # Check if the column is categorical
        df[col] = df[col].fillna(df[col].mode()[0])  # Mode imputation
    else:
        df[col] = df[col].fillna(df[col].median())  # Median if numerical

# For boolean columns, fill missing values with False
boolean_cols = ['HAPPY', 'NEUTRAL', 'RESTED/RELAXED',
                'SAD', 'TENSE/ANXIOUS', 'TIRED']

for col in boolean_cols:
    df[col] = df[col].fillna(False)

# Check for any remaining missing values
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])
```
