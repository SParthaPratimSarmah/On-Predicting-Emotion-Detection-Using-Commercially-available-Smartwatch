```python
import matplotlib.pyplot as plt

# Data
epochs =    [20, 100,200,300,400,500,600,700,800,900,1000]
train_acc = [82.30, 95.60, 97.36, 97.72, 98.07, 98.03, 98.00, 97.95, 98.04, 98.15, 98.11]
val_acc =   [80.35, 94.68, 96.82, 97.32, 97.64, 97.69, 97.60, 97.70, 97.74, 97.74, 97.69]
test_acc =  [79.15, 93.85, 96.20, 96.53, 96.96, 96.99, 97.00, 96.82, 96.85, 96.97, 96.90]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(epochs, train_acc, 'r', label='Training Accuracy')
plt.plot(epochs, val_acc, 'g', label='Validation Accuracy')
plt.plot(epochs, test_acc, 'b', label='Testing Accuracy')

plt.title('Training, Validation, and Testing Accuracy over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')

# Show all epochs on x-axis, rotate for better readability
plt.xticks(epochs, rotation=45, ha='right')

plt.legend()
plt.grid(True)
plt.tight_layout()  # Adjust layout to prevent cutoff of labels
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
models = ['SVM', 'BiLSTM', 'Logistic Regression', 'Decision Tree', 'LightGBM', 'XGBoost', 'Random Forest']
accuracy = [72.56, 73.03, 74.11, 74.92, 76.40, 78.41, 78.75]

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(models, accuracy, color=['red', 'green', 'blue', 'purple', 'orange', 'cyan', 'magenta'])

# Adding labels and title
plt.xlabel('Model Name')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy of Different Models')

# Show values on top of bars
for i, v in enumerate(accuracy):
    plt.text(i, v + 0.2, f"{v}%", ha='center', fontsize=10)

plt.ylim(70, 80)  # Set y-axis limit for better visual comparison
plt.tight_layout()

# Show plot
plt.show()
```
